```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
%matplotlib notebook

#first the imports 

import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np
import time

x=np.linspace(-50., 50., 1000)
sigma_0=1 #Initial standard deviation
D=10 #Diffusion coefficient of Brownian motion
lamb=2 #Imaginary part of Diffusion coefficient of SE equation


fig = plt.figure()
ax = fig.add_subplot(111)
plt.ion()

fig.show()
fig.canvas.draw()

for i in range(0,80):
    #print i
    ax.clear()
    #ax.plot(matrix[i,:])
    CL_sigma_t=np.sqrt(sigma_0**2+2*D*(i/5.0))
    MQ_sigma_t=np.sqrt(sigma_0**2+lamb**2*(i/5.0)**2/sigma_0**2)
    plt.plot(x, 1/(CL_sigma_t * np.sqrt(2 * np.pi)) * np.exp( - (x)**2 / (2 * CL_sigma_t**2) ), linewidth=2, color='r',label='Classic pdf')
    plt.plot(x, 1/(MQ_sigma_t * np.sqrt(2 * np.pi)) * np.exp( - (x)**2 / (2 * MQ_sigma_t**2) ), linewidth=2, color='b',label='Quantum pdf')
    plt.plot(0,0,label="time t = {}".format(i/5.0))
    plt.xlabel('x')
    plt.ylabel('$p(x,t)$')
    plt.legend(loc='upper left')
    plt.ylim((0,0.5))
    plt.show()
    fig.canvas.draw()
    time.sleep(0.5)
```

```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
%matplotlib notebook

#first the imports 

import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np
import time
from scipy import special

x=np.linspace(-50., 50., 1000)
sigma_0=1 #Initial standard deviation
D=10 #Diffusion coefficient of Brownian motion
lamb=2 #Imaginary part of Diffusion coefficient of SE equation


fig = plt.figure()
ax = fig.add_subplot(111)
plt.ion()

fig.show()
fig.canvas.draw()

for i in range(0,80):
    #print i
    ax.clear()
    #ax.plot(matrix[i,:])
    CL_sigma_t=np.sqrt(sigma_0**2+2*D*(i/5.0))
    MQ_sigma_t=np.sqrt(sigma_0**2+lamb**2*(i/5.0)**2/sigma_0**2)
    plt.plot(x, x, linewidth=2, color='r')
    plt.plot((1/2.0)*(1+special.erf(x/(np.sqrt(2)*CL_sigma_t))), (1/2.0)*(1+special.erf(x/(np.sqrt(2)*MQ_sigma_t))), linewidth=2, color='b',label="Quantum transf. at time t = {}".format(i/5.0))
    plt.legend(loc='upper left')
    plt.xlabel('p: Classical cdf values')
    plt.ylabel('$\pi(p)$: Quantum cdf')
    plt.ylim((0,1))
    plt.xlim((0,1))
    plt.show()
    fig.canvas.draw()
    time.sleep(0.5)
```
